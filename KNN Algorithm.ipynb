* KNN algorithm is a supervised Machine Learning Algorithm which can be used for both classification and         predictive problems.

**There are 3 categories of Learning Algorithms**

    1) Lazy Learning algorithm : KNN is a lazy learning algorithm because it does not have specialized                training phase or model and uses all the data for training while classification
    
    2) Non-Parametric Learning Algorithm: KNN is also a non parametric learning algorithm because it does not        assumes anything about the underlying data 
    
    3) Eager learners algorithm : learners when given a set of training tuple will construct a generalization        model before receiving new (test) tuple to classify

## Implementation

1) Step 1: load the data 

2) Step 2: we need to choose the value of K (i.e the nearest data point) .. Here K can be any integer

3) For each point in the test data we need to do the following calculations


    1) Calculate the distance between the test data and each row of training data with the help of a                  mathematical method such as (Euclidean, Manhattan or Hamming Distance)
    
    2) Based on the distance sort them in ascending order.
    
    3) It will chose the top K rows from the sorted array.
    
    4) Now it will assign to the test point based on most frequent class of these rows

```python
k_range = range(1,41)
score = []

for item in k_range:
    knn = KNeighborsClassifier(n_neighbors=item)
    knn.fit(X_train,y_train)
    score.append(knn.score(X_test,y_test))
    
plt.scatter(k_range,score)
plt.xlabel('K Range')
plt.ylabel('Accuracy Score')
plt.show()
```
