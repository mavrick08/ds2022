```python
all_info = []
with open('/home/punit/Downloads/Python-Batch/Data-Science-June/test.html') as data:
    content = data.read()
    soup = BeautifulSoup(content,'lxml')
    all_divs = soup.find_all('div',class_='card')
    for item in all_divs:
        course_name = item.find('h5',class_='card-title').text
        course_desc = item.find('p',class_='card-text').text
        course_price = item.find('a',class_='btn btn-primary').text.split(' ')[-1]
        
        my_course_info = {
            "Course Name":course_name,
            "Course Description": course_desc,
            "Course Price":course_price
        }
        
        all_info.append(my_course_info)
        
print(all_info)
```

## example 2

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
html_data = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=')
```

```python
content = html_data.text
```

```python
soup = BeautifulSoup(content,'lxml')
```

```python
company_name = soup.find('h3',class_='joblist-comp-name').text.strip()
experience = soup.find('ul',class_='top-jd-dtl clearfix')
location = experience.find('span').text
experience = experience.li.text.replace('card_travel','')
job_desc = soup.find('ul',class_='list-job-dtl clearfix')
job_description = job_desc.li.text
key_skills = job_desc.span.text.strip().replace(' ,',',')
job_link = soup.find('h2').a['href']

print(f'Company Name is {company_name}')
print(f'Experience is {experience}')
print(f'Job Description is {job_description}')
print(f'Key Skills is {key_skills}')
print(f'Job Link is {job_link}')
```

```python
job_list = []
all_company_name = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
for item in all_company_name:
    company_name = item.find('h3',class_='joblist-comp-name').text.strip()
    experience = item.find('ul',class_='top-jd-dtl clearfix')
    location = experience.find('span').text
    experience = experience.li.text.replace('card_travel','')
    job_desc = item.find('ul',class_='list-job-dtl clearfix')
    job_description = job_desc.li.text
    key_skills = job_desc.span.text.strip().replace(' ,',',')
    job_link = item.find('h2').a['href']
    
    
    my_job_dict  = {
        "Company Name":company_name,
        "Experience":experience,
        "Location":location,
        "Job Desc":job_description,
        "Key Skill":key_skills,
        "Job Link":job_link
    }
    
    job_list.append(my_job_dict)
    
print(job_list)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(job_list)
```

```python
df
```

```python
df.to_csv('My-Python-Job.csv')
```

## E-commerce scrapping 

```python
html_data = requests.get('https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page=1')
content = html_data.text
soup = BeautifulSoup(content,'lxml')
```

```python
product_name = soup.find('div',class_='caption').a.text
product_link = soup.find('div',class_='caption').a['href']
product_link = 'https://webscraper.io'+product_link
product_price = soup.find('h4',class_='pull-right price').text
product_description = soup.find('p',class_='description').text
review = soup.find('div',class_='ratings').p.text
```

```python
page_1_product = soup.find_all('div',class_='col-sm-4 col-lg-4 col-md-4')

for items in page_1_product:
    product_name = items.find('div',class_='caption').a.text
    product_link = items.find('div',class_='caption').a['href']
    product_link = 'https://webscraper.io'+product_link
    product_price = items.find('h4',class_='pull-right price').text
    product_description = items.find('p',class_='description').text
    review = items.find('div',class_='ratings').p.text
    
    my_prod_dict = {
        'Product Name':product_name,
        'Product Price':product_price,
        'Product Description':product_description,
        'Product Link':product_link,
        'Review':review
    }
    
    print(my_prod_dict)
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page='
```

```python
lst_url = []
for value in range(1,21):
    lst_url.append(url+str(value))
```

```python
for val in lst_url:
    print(val)
```

```python
prod_list = []
def get_all_product_data():
    for link in lst_url:
        response = requests.get(link)
        html_text = response.text
        product_soup = BeautifulSoup(html_text,'lxml')
        product_list = product_soup.find_all('div',class_='col-sm-4 col-lg-4 col-md-4')
        for item in product_list:
            product_name = item.find('div',class_='caption').a.text
            product_link = item.find('div',class_='caption').a['href']
            product_link = 'https://webscraper.io'+product_link
            product_price = item.find('h4',class_='pull-right price').text
            product_description = item.find('p',class_='description').text
            review = item.find('div',class_='ratings').p.text
            my_prod_dict = {
            'Product Name':product_name,
            'Product Price':product_price,
            'Product Description':product_description,
            'Product Link':product_link,
            'Review':review
            }
            
            prod_list.append(my_prod_dict)
get_all_product_data()
```

```python
prod_df = pd.DataFrame(prod_list)
prod_df
```

```python
prod_df.to_csv('Laptop-Product-List.csv')
```
