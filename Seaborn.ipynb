```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
help(sns.load_dataset)
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df
```

```python
sns.set(style='whitegrid')
```

```python
fig = plt.figure()
fig.set_figwidth(12)
fig.set_figheight(7)
sns.relplot(x='total_bill',y='tip',data=tips_df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips_df,hue='sex')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips_df,hue='time')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips_df,hue='size',palette='brg')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips_df,palette='Reds',hue='sex')
```

```python
sns.catplot(x='day',y='total_bill',data=tips_df)
```

```python
sns.catplot(x='day',y='total_bill',data=tips_df,jitter=False)
```

```python
sns.catplot(x='day',y='total_bill',data=tips_df,kind='swarm',hue='sex')
```

```python
sns.histplot(data=tips_df,x='total_bill',color='red')
```

```python
sns.histplot(data=tips_df,x='day',hue='sex',multiple='dodge',shrink=.9,palette='Reds')
```

```python
sns.distplot(tips_df['total_bill'],kde=True,color='red')
```

```python
sns.distplot(tips_df['total_bill'],hist=False,color='red')
```

```python
sns.countplot(x='sex',data=tips_df,hue='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips_df)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips_df)
```
